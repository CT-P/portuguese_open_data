```python
# Import Libraries
from textblob import TextBlob
import sys
import tweepy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

#import pycountry
import re
import string
import logging
import time
```

```python
# Authentication
consumerKey = ...
consumerSecret = ...
accessToken = ...
accessTokenSecret = ...
auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
df_twitter=pd.read_csv('input_data/twitter_deputies.csv')
df_twitter.twitter=df_twitter.twitter.astype('str')
df_twitter.head(3)
```

```python

tweetsC=[]
for us in df_twitter.twitter:
    if  us!='nan':
        print(us)
        ini_date=str(df_twitter[df_twitter.twitter==us].year_ini.values[0])+'01010000'
        end_date=str(df_twitter[df_twitter.twitter==us].year_max.values[0])+'12300000'
        try:
            tweets=tweepy.Cursor(api.search_full_archive,label='opendatapt', query='from:'+us+' -is:reply -is:retweet', fromDate=ini_date, toDate=end_date).items(3000)
            #tweetsC.append([tweets.user.screen_name,tweets.text, tweets.created_at.year, tweets.id,"https://twitter.com/" + str(tweets.user.screen_name) + "/status/" + str(tweets.id)])
                                        
            
                
            #tweet = tweets.next()
            tweetsC.append([[tweet.user.screen_name,tweet.text, tweet.created_at.year, tweet.id,"https://twitter.com/" + str(tweet.user.screen_name) + "/status/" + str(tweet.id)] for tweet in tweets])
        except StopIteration:
            print(len(tweetsC))
            break #return
        except tweepy.errors.TweepyException as e:
            #if 'Failed to send request:' in e.reason:
            print ("Time out error caught.", e)
            time.sleep(180)
            continue

flat_list = [item for sublist in tweetsC for item in sublist]

df=pd.DataFrame(flat_list)
df.columns=['user_name', 'text', 'year', 'tweet_id', 'url']
df.to_pickle('twitter_scrapping/raw_tweets.pkl')
```

```python
repliesC=[]
for us in df_twitter.twitter:
    if  us!='nan':
        print(us)
        
        ini_date=str(df_twitter[df_twitter.twitter==us].year_ini.values[0])+'01010000'
        end_date=str(df_twitter[df_twitter.twitter==us].year_max.values[0])+'12300000'
        replies=tweepy.Cursor(api.search_full_archive,label='opendatapt', query='to:'+us+' is:reply -is:retweet', fromDate=ini_date, toDate=end_date).items(3000)
        try:
            
            repliesC.append([[us,reply.text, reply.created_at.year, reply.id,"https://twitter.com/" + str(reply.user.screen_name) + "/status/" + str(reply.id)] for reply in replies])
        except StopIteration:
            print(len(repliesC))
            break #return
        except tweepy.errors.TweepyException as e:
            #if 'Failed to send request:' in e.reason:
            print ("Time out error caught.", e)
            time.sleep(180)
            continue

Rflat_list = [item for sublist in repliesC for item in sublist] 
dfR=pd.DataFrame(Rflat_list)
dfR.columns=['user_name', 'text', 'year', 'tweet_id', 'url']
dfR.to_pickle('twitter_scrapping/raw_replies.pkl')
```
